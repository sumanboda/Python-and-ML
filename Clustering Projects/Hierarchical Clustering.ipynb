```python
# importing all the essential libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing the data set
dataset = pd.read_csv('Mall_Customers.csv')
print(dataset)
print()
x=dataset.iloc[:,[3,4]].values
print(x)  # here havinh genderand age column doesnt effect
```

```python
# using the dendrograms method find the optimun number of clusters
import scipy.cluster.hierarchy as sch
dend=sch.dendrogram(sch.linkage(x,method="ward"))
#linkage is an algorithm in HC metod ward means based on variance vd in the clusters it is similar to wcss
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
# from the figure the largest vertical line where it doesnot cross any horizontal line ..
# now if we draw any horizontal line from that largest vertical line it crosses the 5 vertical lines so 5 clusters
```

```python
# # Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
y_hc=hc.fit_predict(x)
y_hc
```

```python
## Visualising the clusters
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
